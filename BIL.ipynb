```python
timesData = pd.read_csv('cwurData.csv')
print(timesData.shape)
timesData.head()
```

```python
df =  timesData.iloc[:]
df2012 = timesData[timesData.year == 2012].iloc[:100,:]
df2013 = timesData[timesData.year == 2013].iloc[:100,:]
df2014 = timesData[timesData.year == 2014].iloc[:100,:]
df2015 = timesData[timesData.year == 2015].iloc[:100,:]

# creating trace1
trace1 =go.Scatter(
                    x = df2012.world_rank,
                    y = df2012.score,
                    mode = "markers",
                    name = "2012",
                    text= df2012.institution)
# creating trace2
trace2 =go.Scatter(
                    x = df2013.world_rank,
                    y = df2013.score,
                    mode = "markers",
                    name = "2013",
                    text= df2013.institution)
# creating trace3
trace3 =go.Scatter(
                    x = df2014.world_rank,
                    y = df2014.score,
                    mode = "markers",
                    name = "2014",
                    text= df2014.institution)
# creating trace4
trace4 =go.Scatter(
                    x = df2015.world_rank,
                    y = df2015.score,
                    mode = "markers",
                    name = "2015",
                    text= df2015.institution)
data = [trace1, trace2, trace3, trace4]
layout = dict(title = 'Score vs world rank of top 100 universities with 2012, 2013, 2014 and 2015 years',
              xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Score',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

```python
X = np.array(list(zip(df.world_rank,df.score)))
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=3)
# Fitting the input data
kmeans = kmeans.fit(X)
# Getting the cluster labels
labels = kmeans.predict(X)
# Centroid values
centroids = kmeans.cluster_centers_
print(centroids)
```
